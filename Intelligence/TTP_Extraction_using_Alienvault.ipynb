```python
import pandas as pd

# Read the dataset into a pandas DataFrame
Tor_SRC_df = pd.read_csv('/content/drive/MyDrive/xai/Darknet/AV/SRC_Tor.csv')
VPN_SRC_df = pd.read_csv('/content/drive/MyDrive/xai/Darknet/AV/SRC_VPN.csv')

Non_Tor_SRC_df = pd.read_csv('/content/drive/MyDrive/xai/Darknet/AV/SRC_NoNTOR.csv')
Non_VPN_SRC_df = pd.read_csv('/content/drive/MyDrive/xai/Darknet/AV/SRC_Non_VPN.csv')

Tor_DST_df = pd.read_csv('/content/drive/MyDrive/xai/Darknet/AV/DST_Tor.csv')
VPN_DST_df = pd.read_csv('/content/drive/MyDrive/xai/Darknet/AV/DST_VPN.csv')

Non_Tor_DST_df = pd.read_csv('/content/drive/MyDrive/xai/Darknet/AV/DST_Non_Tor.csv')
Non_VPN_DST_df = pd.read_csv('//content/drive/MyDrive/xai/Darknet/AV/DST_Non_VPN.csv')
```

```python
# Convert 'Pulse Targeted Countries' values to strings and replace nan with an empty string
import numpy as np

# Assuming combined_df is your DataFrame

Non_Tor_SRC_df['Pulse Attack ID'] = Non_Tor_SRC_df['Pulse Attack ID'].apply(lambda x: str(x) if pd.notna(x) else '[]')

Non_Tor_DST_df['Pulse Attack ID'] = Non_Tor_DST_df['Pulse Attack ID'].apply(lambda x: str(x) if pd.notna(x) else '[]')
```

```python
import pandas as pd

# Assuming df1 and df2 are your two DataFrames
# Concatenate the DataFrames
combined_df = pd.concat([Tor_SRC_df, VPN_SRC_df,Non_Tor_SRC_df,Non_VPN_SRC_df,Tor_DST_df,VPN_DST_df,Non_Tor_DST_df,Non_VPN_DST_df], ignore_index=True)

# Assuming df is your DataFrame and 'column_name' is the column based on which you want to remove duplicates
combined_df = combined_df.drop_duplicates(subset='IP Address', keep='first')

# If you want to reset the index after removing duplicates
combined_df = combined_df.reset_index(drop=True)
```

```python
# Assuming combined_df is your DataFrame
combined_df['Pulse Targeted Countries'] = combined_df['Pulse Targeted Countries'].astype(str).replace('nan', '[]')
combined_df['Pulse Attack ID'] = combined_df['Pulse Attack ID'].astype(str).replace('nan', '[]')
```

```python
attack_id_df = combined_df[combined_df['Pulse Attack ID'].apply(lambda x: isinstance(x, str) and x != '[]')]
```

```python
attack_id_df = attack_id_df.dropna(subset=['Pulse Attack ID'])
# Merge all attack IDs into a single variable without removing duplicates
combined_attack_ids = ', '.join(attack_id_df['Pulse Attack ID'].astype(str))

# Split the string into a list of attack IDs and clean up the formatting
attack_id_list = [id.split('.')[0].strip("[]'") for id in combined_attack_ids.split(', ')]
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from collections import Counter

# Count the frequency of each word
word_freq = Counter(attack_id_list)

# Generate the word cloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(word_freq)

# Plot the word cloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
# Save as PNG
plt.savefig('/content/drive/MyDrive/xai/Darknet/AV/img/Attack_Technique.png', format='png', bbox_inches='tight')

# Save as EPS
plt.savefig('/content/drive/MyDrive/xai/Darknet/AV/img/Attack_Technique.eps', format='eps', bbox_inches='tight')
plt.show()
```
